### функции исправления

```python
def corrected_rooms(df):
    df.loc[(df['Rooms'] == 0) & (df['Square'] < 56), 'Rooms'] = 1.0
    df.loc[(df['Rooms'] == 0) & (df['Square'] > 55) & (df['Square'] < 76), 'Rooms'] = 2.0
    df.loc[(df['Rooms'] == 0) & (df['Square'] > 75) & (df['Square'] < 96), 'Rooms'] = 3.0
    df.loc[(df['Rooms'] == 0) & (df['Square'] > 115) & (df['Square'] < 167), 'Rooms'] = 5.0
    df.loc[(df['Rooms'] == 0) & (df['Square'] > 167), 'Rooms'] = 6.0
    return df
```

```python
data = corrected_rooms(data)
```

```python
test = corrected_rooms(test)
```

```python
data['LifeSquare'] = data['LifeSquare'].fillna(25)
```

```python
def corrected_square(df):
    df.loc[df['Square'] < 25, 'Square'] = 25
```

```python
corrected_square(data)
```

```python
corrected_square(test)
```

```python
def corrected_lifesquare(df):
    df.loc[(df['Square'] // 2) > df['LifeSquare'], 'LifeSquare'] = df['Square'] // 2
    return df
```

```python
data = corrected_lifesquare(data)
```

```python
test = corrected_lifesquare(test)
```

```python
def corrected_kitchen(df):
    df.loc[df['KitchenSquare'] < 5, 'KitchenSquare'] = 5
```

```python
corrected_kitchen(data)
```

```python
corrected_kitchen(test)
```

```python
def corrected_floor(df):
    df.loc[df['Floor'] > df['HouseFloor'], 'HouseFloor'] = df['Floor']
```

```python
corrected_floor(data)
```

```python
corrected_floor(test)
```
